```python
import numpy as np
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 1000)
#tenosus, et Y<-4
prob_normal = norm.cdf(-4)
prob_student = t.cdf(-4, df=5)

plt.figure(figsize=(8, 6))

normal_pdf = norm.pdf(X)
student_pdf = t.pdf(X, df=5)
plt.plot(X, normal_pdf, label='Normaaljaotus', linewidth=2)
plt.plot(X, student_pdf, '--', label="Student'i t-jaotus (df=5)", linewidth=2)

plt.axvline(x=-4, color='r', linestyle=':', label='X = -4')

plt.title("Normaal- vs Student'i t-jaotus: \n tenosus, et X<-4")
plt.xlabel('X')
plt.ylabel('Tenosustihedus')

plt.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=2)
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

plt.grid(True)
plt.show()
print(f"Suhe (Student'i-t/Normaal): {prob_student/prob_normal:.2f}x")
```
